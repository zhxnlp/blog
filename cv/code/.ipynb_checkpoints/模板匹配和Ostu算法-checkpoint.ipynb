```python
import cv2
import numpy as np

img = cv2.imread('./lena.jpg', 0)
template = cv2.imread('./face.jpg', 0)
print(img.shape)
print(template.shape)
# 匹配完之后的结果的大小应该是 263 - 110 + 1 , 263 - 85 + 1
res = cv2.matchTemplate(img, template, cv2.TM_SQDIFF)
print(res.shape)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
print(min_val)
print(max_val)
print(min_loc)
print(max_loc)

# 画出匹配到的位置
img = cv2.rectangle(img, min_loc, (min_loc[0] + 85, min_loc[1] + 110), (0, 0, 255), 2)

cv2.imshow('tempalte', template)
cv2.imshow('img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 6种匹配方式的对比

```python
eval('cv2.TM_CCOEFF')
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('./lena.jpg', 0)
template = cv2.imread('./face.jpg', 0)

# 定义六种方法的列表
methods = ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR', 'cv2.TM_CCORR_NORMED', 
          'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']

w, h = template.shape[1], template.shape[0]
for method in methods:
    img2 = img.copy()
    
    # 匹配
    res = cv2.matchTemplate(img, template, eval(method))
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
    
    if method in ['cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']:
        top_left = min_loc
    else:
        top_left = max_loc
        
    bottom_right = (top_left[0] + w, top_left[1] + h)
    
    # 画矩形
    cv2.rectangle(img2, top_left, bottom_right, (0, 0, 255), 2)
    # 画子图, 都画在一起. 
    plt.subplot(121), plt.imshow(res, cmap='gray')
    plt.axis('off')
    plt.subplot(122), plt.imshow(img2, cmap='gray')
    plt.axis('off')
    plt.title(method)
    plt.show()
```

### 匹配多个对象

```python
import cv2
import numpy as np

mario = cv2.imread('./mario.jpg')
# 灰度化
mario_gray = cv2.cvtColor(mario, cv2.COLOR_BGR2GRAY)
template = cv2.imread('./mario_coin.jpg', 0)
h, w = template.shape[:2]

res = cv2.matchTemplate(mario_gray, template, cv2.TM_CCOEFF_NORMED)
# 设定阈值, 认为相关系数大于0.8, 就认为是匹配上了.
threshold = 0.8
# 分别返回的是y轴, x轴的索引
loc = np.where(res >= threshold)
# np.argwhere(res >=threshold)
# zip(np.array([40, 40, 40, 40, ...]), np.array(69, 70, 83, ...))
for pt in zip(*loc[::-1]):
    bottom_right = (pt[0] + w, pt[1] + h)
    cv2.rectangle(mario, pt, bottom_right, (0, 0, 255), 2)
    
cv2.imshow('mario', mario)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### ostu算法

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

naza = cv2.imread('./naza.png')
naza_gray = cv2.cvtColor(naza, cv2.COLOR_BGR2GRAY)

# ostu特别适合双峰结构的灰度图
_ = plt.hist(naza_gray.ravel(), bins=256, range=[0, 255])

# 普通阈值处理
ret, dst = cv2.threshold(naza_gray, 180, 255, cv2.THRESH_BINARY)
cv2.imshow('naza', np.hstack((naza_gray, dst)))

# 使用Otsu算法
ret, dst = cv2.threshold(naza_gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print(ret)
cv2.imshow('ostu', np.hstack((naza_gray, dst)))


cv2.waitKey(0)
cv2.destroyAllWindows()
```
