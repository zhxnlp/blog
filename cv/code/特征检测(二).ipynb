```python
import cv2
import numpy as np

img = cv2.imread('chess.png')
# img = cv2.imread('chess.png', flags=cv2.IMREAD_GRAYSCALE)

# 灰度化
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# shi-tomasi焦点检测
corners = cv2.goodFeaturesToTrack(gray, maxCorners=1000, qualityLevel=0.1, minDistance=10)
# print(corners)
# print(type(corners))
print(corners.shape)
corners = np.int0(corners)
# img[dst>0.01 * dst.max()]

# 画出角点
for i in corners:
    # i相当于corners中的每一行数据
    # ravel()把二维变一维了.即角点的坐标点
    x,y = i.ravel()
    cv2.circle(img, (x, y), 3, (0, 0, 255), -1)
    
cv2.imshow('Shi-Tomasi', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### sift算法

```python
import cv2
import numpy as np

img = cv2.imread('chess.png')
# img = cv2.imread('chess.png', flags=cv2.IMREAD_GRAYSCALE)

# 灰度化
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 创建sift对象
sift = cv2.SIFT_create()
# sift算法已经专利过期了.
# cv2.SIFT_create()

# 进行检测
kp = sift.detect(gray)

# 计算描述子
kp, des = sift.compute(gray, kp)
# kp是一个列表, 里面存放的是封装的KeyPoint对象.
# 还可以 一步到位计算
kp, des = sift.detectAndCompute(gray, None)
# print(kp)
print(len(kp))
print(type(des))
print(des.shape)
print(des[0])

# 绘制关键点
cv2.drawKeypoints(gray, kp, img)
cv2.imshow('SIFT', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
